# 06 — Experiment D: Quantum Model Wildfire Risk
Uses wildfire risk probabilities from Shreyas's Quantum Kernel ZZ-6q model
(ROC-AUC 0.622 — best quantum model).

File: `data/external/wildfire_risk_quantum_2021.csv`

This experiment answers the core challenge question:
**Do quantum-generated wildfire risk scores improve insurance premium prediction?**

```python
import sys
sys.path.append('..')
import pandas as pd
import numpy as np
from src.preprocessing import load_data, drop_leakage_columns, fix_target, encode_booleans, split_train_test, merge_wildfire_risk
from src.features import build_all_features
from src.models import run_experiment
from src.utils import save_metrics
from src.config import BASE_FEATURES, LAG_FEATURES, TARGET, QUANTUM_RISK_FILE, QUANTUM_RISK_COL

EXPERIMENT_NAME = 'D - Quantum Wildfire Risk'
FEATURES_D = BASE_FEATURES + LAG_FEATURES + [QUANTUM_RISK_COL]

df = load_data()
df = drop_leakage_columns(df)
df = fix_target(df)
df = encode_booleans(df)
df = build_all_features(df)
df = merge_wildfire_risk(df, QUANTUM_RISK_FILE, QUANTUM_RISK_COL)
train, test = split_train_test(df)

available = [f for f in FEATURES_D if f in df.columns]
missing_f = [f for f in FEATURES_D if f not in df.columns]
if missing_f:
    print(f'Warning: {len(missing_f)} features not found: {missing_f}')
print(f'Using {len(available)} features')

metrics_D = run_experiment(train, test, available, EXPERIMENT_NAME)
results_D = save_metrics(metrics_D, 'experiment_D_metrics.csv')
print(results_D.to_string(index=False))
print('✅ Experiment D complete')
```
