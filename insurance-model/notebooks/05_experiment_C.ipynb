# 05 — Experiment C: Classical Model Wildfire Risk
Uses wildfire risk probabilities from Shreyas's Logistic Regression model
(strongest classical model, ROC-AUC 0.865).

File: `data/external/wildfire_risk_classical_2021.csv`

```python
import sys
sys.path.append('..')
import pandas as pd
import numpy as np
from src.preprocessing import load_data, drop_leakage_columns, fix_target, encode_booleans, split_train_test, merge_wildfire_risk
from src.features import build_all_features
from src.models import run_experiment
from src.utils import save_metrics
from src.config import BASE_FEATURES, LAG_FEATURES, TARGET, CLASSICAL_RISK_FILE, CLASSICAL_RISK_COL

EXPERIMENT_NAME = 'C - Classical Wildfire Risk'
FEATURES_C = BASE_FEATURES + LAG_FEATURES + [CLASSICAL_RISK_COL]

df = load_data()
df = drop_leakage_columns(df)
df = fix_target(df)
df = encode_booleans(df)
df = build_all_features(df)
df = merge_wildfire_risk(df, CLASSICAL_RISK_FILE, CLASSICAL_RISK_COL)
train, test = split_train_test(df)

available = [f for f in FEATURES_C if f in df.columns]
missing_f = [f for f in FEATURES_C if f not in df.columns]
if missing_f:
    print(f'Warning: {len(missing_f)} features not found: {missing_f}')
print(f'Using {len(available)} features')

metrics_C = run_experiment(train, test, available, EXPERIMENT_NAME)
results_C = save_metrics(metrics_C, 'experiment_C_metrics.csv')
print(results_C.to_string(index=False))
print('✅ Experiment C complete')
```
